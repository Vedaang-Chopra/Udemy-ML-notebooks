```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("../MISSION Machine Learning Datasets/Churn_Modelling.csv")
```

```python
dataset.head()
```

```python
X = dataset.iloc[:,3:13].values
Y = dataset.iloc[:,13].values
```

```python
#encoding categorical values
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_encoder_X_1 = LabelEncoder()
X[:,1] = label_encoder_X_1.fit_transform(X[:,1])
label_encoder_X_2 = LabelEncoder()
X[:,2] = label_encoder_X_2.fit_transform(X[:,2])
# now as 1 colum of X has more then 2 categorical variables therefore we need to create the dummies
onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
#now to avoid dummy trap let's remove one dummy variable column
X = X[:,1:]
```

```python
X = pd.DataFrame.from_records(X)
X
```

```python
X = X.values
```

```python
X
```

```python
# Splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 0)
```

```python
# Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow
```

```python
from tensorflow import keras
```

```python
## importing keras libraries and packages
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# initializing ANN
classifier = Sequential()
```

```python
# adding the input layer and first hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
```

```python
#adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
```

```python
#adding the output layer
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

```python
#compiling ANN
classifier.compile(optimizer='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# fitting the ANN to training set
classifier.fit(X_train, Y_train,batch_size= 10, epochs = 100)
```

```python
#getting predictions
Y_pred = classifier.predict(X_test)
```

```python
Y_pred
```

```python
Y_pred = (Y_pred > 0.5)
```

```python
Y_pred
```

```python
#confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_pred)
```
