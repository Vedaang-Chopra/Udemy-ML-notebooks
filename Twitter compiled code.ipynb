```python
import pandas as pd
import json
dat=list()
with open ('../MISSION Machine Learning Datasets/midha.json', 'r') as f:
    for l in f.readlines():
        if not l.strip (): # skip empty lines
            continue

        json_data = json.loads (l)
        dat.append(json_data)
df = pd.DataFrame(dat)
```

```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range (0,1183):
    tweet = re.sub('[^a-zA-Z]', ' ', df['text'][i])
    tweet = tweet.lower()
    tweet = tweet.split()
    s = PorterStemmer()
    tweet = [s.stem(word) for word in tweet if not word in set(stopwords.words('english'))]
    tweet = ' '.join(tweet)
    corpus.append(tweet)
```

```python
## creating bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1000)
X = cv.fit_transform(corpus).toarray()
```
